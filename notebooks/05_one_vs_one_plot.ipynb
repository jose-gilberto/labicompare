```python
import pandas as pd
from labicompare.visualization import one_vs_one_plot
```

### Loading the metrics dataset

Unlikely the critical difference plot, here we don't need to have a fixed format for the metrics dataset. To use the One-versus-One plot you only need two different arrays of the same length, containing the metrics for the two compared models on all datasets.

For simplicity we still use the metrics dataset as a base, but you can modify this cell to use whatever data structure you want to use, the only necessary thing is that they must have the same length and be a `numpy.ndarray`.

```python
metrics = pd.read_csv('./example_metrics.csv')
metrics
```

### Generating the One-versus-One Plot

Once you have the metrics loaded in the compatible format, you only have to call the function `one_vs_one_plot` and pass the metrics arrays, as well as the model names to render the following plot:

```python
one_vs_one_plot(
    metrics_model_1=metrics['LR'].values,
    metrics_model_2=metrics['LDA'].values,
    model_1_name='LogisticReg',
    model_2_name='LDA'
)
```
